```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout

!git clone https://github.com/hmyrcmn/KUL.git

# Veri yükleme
true_folder ='/content/KUL/coordinateds2/trueValues'
X_data = []
Y_labels = []

for filename in os.listdir(true_folder):
    data = np.genfromtxt(os.path.join(true_folder, filename), delimiter=',')
    # X_data.append(data)
    X_data.append(data[:, 1:])  # Sadece virgül sonrasındaki değerleri ekledim

    Y_labels.append(1)








false_folder = '/content/KUL/coordinateds2/falseValues'
for filename in os.listdir(false_folder):
    data = np.genfromtxt(os.path.join(false_folder, filename), delimiter=',')
    # X_data.append(data)
    X_data.append(data[:, 1:])  # Sadece virgül sonrasındaki değerleri ekledim

    Y_labels.append(0)




X_data = np.array(X_data)
Y_labels = np.array(Y_labels)
print("x_data shape: ",X_data.shape)
X_data
```

# sclaing

```python
# Verileri düzenleme ve normalleştirme
num_samples = len(X_data)
sequence_length = X_data.shape[1]
num_features = X_data.shape[2]

X_data_normalized = X_data.reshape(num_samples, sequence_length * num_features)

# Eğitim ve test setlerine bölme


X_train, X_test, y_train, y_test = train_test_split(X_data_normalized, Y_labels, test_size=0.2, random_state=42)

print("X_train.shape:",X_train.shape)
X_train
```

# train model

Stochastic Gradient Descent (SGD): Temel optimizasyon algoritmasıdır. Her eğitim örneği üzerinden bir gradient hesaplar ve model parametrelerini günceller.

Adam (Adaptive Moment Estimation): Adaptif bir öğrenme oranına sahip olan ve momentum terimini içeren popüler bir optimizasyon algoritmasıdır. Adam, özellikle çeşitli uygulamalarda iyi performans gösterir.

RMSprop (Root Mean Square Propagation): Öğrenme oranını adapte eden bir başka popüler optimizasyon algoritmasıdır. Quadratic bir şekilde birikmiş kareli gradientler kullanılarak öğrenme oranını günceller.

Adagrad (Adaptive Gradient Algorithm): Parametre spesifik öğrenme oranları kullanarak güncelleme yapar. Daha önce görülmüş örneklerin daha düşük öğrenme oranlarına sahip olmasını sağlar.

Adadelta: Adagrad'ın bir genelleştirmesidir ve adaptif öğrenme oranları kullanır. Ancak, Adadelta'nın Adagrad'a kıyasla daha az hiperparametresi vardır.

Nadam: Nesterov Accelerated Gradient (NAG) ve Adam algoritmalarının bir kombinasyonudur. NAG, momentum terimini kullanarak hızlı konverjans sağlar.

FTRL (Follow The Regularized Leader): Çeşitli regülasyon terimlerini içeren bir optimizasyon algoritmasıdır. Özellikle büyük ölçekli çevrim içi öğrenme uygulamalarında kullanılır.
